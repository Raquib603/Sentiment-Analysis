```python
from transformers import pipeline
```

```python
emotionModel = pipeline ('text-classification', model = 'SamLowe/roberta-base-go_emotions')
```

Taking Input

```python
label = emotionModel(['you should do that'])
label
```

Use it on dataset

```python
import pandas as pd

df = pd.read_csv('/isear.csv')
df.columns = ['Emotions', 'Text']
df.head()
```

```python
df.shape
```

```python
simpleText = df[:200]
simpleText.shape
```

```python
def get_emotion(text):
  return (emotionModel(text)[0]['label'])
```

```python
simpleText['Text'][20:30].apply(get_emotion)
```

```python
simpleText['RoBertaEmotions'] = simpleText['Text'].apply(get_emotion)
```
